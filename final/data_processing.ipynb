```python
import numpy as np
import pandas as pd
filepath="dataset/train_lpd.csv"
data=pd.read_csv(filepath,encoding_errors='ignore')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
data['Gender of the patient']=enc.fit_transform(data['Gender of the patient'])
data.info()
```

```python
data.describe()
```

```python
data=data.fillna(method='bfill')
data.isnull().sum()
```

```python
data=data.drop_duplicates()
data.info()
```

```python
data['Result'].value_counts()
```

```python
data['Gender of the patient'].value_counts()
```

```python
filtered_data = data[data['Result'] == 2]
print(filtered_data)
```

```python
cls_0=data[data['Result']==1]
cls_1=data[data['Result']==2]
```

```python
len(cls_1)
```

```python
cls_1=cls_1.sample(13770,replace=True)
data=pd.concat([cls_0,cls_1],axis=0)
data.info()
```

```python
for x in data.columns:
    data[x]=(data[x]-data[x].min())/(data[x].max()-data[x].min())
data.head()
```
