```python
import pandas as pd
import plotly.express as px

# Load the dataset into a DataFrame
df = pd.read_csv('../dataset/train_lpd.csv',encoding_errors='ignore')

# Choose the column to plot
column_name = 'Age of the patient'  # Replace with the actual column name

# Create the Plotly figure
fig = px.histogram(df, x=column_name)

# Show the plot
fig.show()
```

```python
# Choose the column to plot
column_name = 'Result'  # Replace with the actual column name

# Sort the DataFrame by the column in ascending order
df_sorted = df.sort_values(column_name)

# Create the Plotly figure
fig = px.histogram(df_sorted, x=column_name)
min_value = df_sorted[column_name].min()
max_value = df_sorted[column_name].max()

# Get the statistical measures
mean_value = df_sorted[column_name].mean()
std_value = df_sorted[column_name].std()
quantiles = df_sorted[column_name].quantile([0.25, 0.5, 0.75])
count = df_sorted[column_name].count()
print("Mean: {:.2f}".format(mean_value))
print("Std: {:.2f}".format(std_value))
print("Q1: {:.2f}".format(quantiles[0.25]))
print("Q2: {:.2f}".format(quantiles[0.5]))
print("Q3: {:.2f}".format(quantiles[0.75]))
print("Count: {}".format(count))
# Add annotations for statistical measures
fig.update_layout(
    annotations=[
         dict(
            x=min_value,
            y=0,
            xref="x",
            yref="paper",
            text="Min: {}".format(min_value),
            showarrow=True,
            arrowhead=1,
            ax=-40,
            ay=-40
        ),
        
        dict(
            x=max_value,
            y=0,
            xref="x",
            yref="paper",
            text="Max: {}".format(max_value),
            showarrow=True,
            arrowhead=1,
            ax=-40,
            ay=-40
        )
        
    ]
)

# Save the figure as an HTML file

# Show the plot
fig.show()
```

```python
fig.write_html('../static/image/result.html')
```
